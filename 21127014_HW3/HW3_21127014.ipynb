## Exercises 1

Use `re.search` to find whether a string contains a phone number. The pattern that you write should detect a phone number in the following strings.  
```
"Call me at 382-384-3840."  
"my number is (510) 849-3519. Call me!"
```  
And not find a match in the following strings. 
```
"my number is 510-849-35192"  
"here’s my number: 510-849.3519"
``` 
Consider making your own tests as well  

````python
phone_number_pattern = r'\b\d{3}[-]\d{3}[-]\d{4}\b|\(\d{3}\)[\s]*\d{3}[-]\d{4}'

test_strings = [
    "Call me at 382-384-3840.",
    "my number is (510) 849-3519. Call me!",
    "my number is 510-849-35192",
    "here’s my number: 510-849.3519",
]

for test_string in test_strings:
    if re.search(phone_number_pattern, test_string):
        print(f"Found : {test_string}")
    else:
        print(f"Not found : {test_string}")
````

## Exercise 2

Use `re.sub` to alter the string below so that the dates have a common format that uses a dash for the day, month, and year separator.  
```
03/12/2018, 03.13.18, 03/14/2018, 03:15:2018
```

````python
# YOUR CODE HERE
text = "03/12/2018, 03.13.18, 03/14/2018, 03:15:2018"

date_pattern = r'(\d{2})[/.:](\d{2})[/.:](\d{4})'

formatted_text = re.sub(date_pattern, r'\1-\2-\3', text)

print(formatted_text)
````

## Exercise 3

Consider the first five sentences of the novel “Little Women” below. Extract the spoken dialog from each sentence.

````python
text = '''
"Christmas won't be Christmas without any presents," grumbled Jo, lying on the rug.
"It's so dreadful to be poor!" sighed Meg, looking down at her old dress.
"I don't think it's fair for some girls to have plenty of pretty things, and other girls nothing at all," added little Amy, with an injured sniff.
"We've got Father and Mother, and each other," said Beth contentedly from her corner.
The four young faces on which the firelight shone brightened at the cheerful words, but darkened again as Jo said sadly, "We haven't got Father, and shall not have him for a long time."
'''
print(text)
````

````python
# YOUR CODE HERE
dialog_pattern = r'"(.*?)"'

spoken_dialog = re.findall(dialog_pattern, text)

for dialog in spoken_dialog:
    print(dialog)
````

In this exercise, you you working with ```email_test.txt``` file (attached), using Regular Expression. 
1. Extract name and email address of sender  
Output should be something like this

<u>**Noted**</u>: If the email don't have name, you can skip this it.
```

```

2. Find SUBJECT of All email with Date from 01/06/2003 to 13/05/2004 </br>
<u>**Noted**</u>:
- If the email don't have Date, you can skip this it.
- SUBJECT may appear before or after Date 

````python
import re

nameSenders = []
mailSenders = []
subject = []

fh = open(r"email_test.txt", "r").read()

emails = re.split(r'(?=\nFrom r)', fh)

# YOUR CODE HERE
# 1. Extract name and email address of sender

def findSenderName(text):
    name_pattern = r'From:\s*"([^"]+)"\s*<[^>]+>'
    name_match = re.search(name_pattern, text)
    sender_name = name_match.group(1).strip() if name_match else "Unknown"
    return sender_name

def findSenderMail(text):
    email_pattern = r'Return-Path:\s*<([^>]+)>'
    email_match = re.search(email_pattern, text)
    sender_email = email_match.group(1) if email_match else "Unknown"
    return sender_email


# 2. Find SUBJECT of All email from 01/06/2003 to 13/05/2004
def findSubject(email_text, start_date_str, end_date_str):
    emails = re.split(r'(?=\n\n)', email_text)

    sub_match = ''
    date_format = "%d %b %Y"

    start_date = datetime.strptime(start_date_str, date_format)
    end_date = datetime.strptime(end_date_str, date_format)

    for email in emails:
        date_match = re.search(r'Date: (.+)', email)
        if date_match:
            date_str = date_match.group(1)

            date_regex = re.search(r'(\d{2} \w{3} \d{4})', date_str)
            if date_regex:
                email_date = datetime.strptime(date_regex.group(1), date_format)

                if start_date <= email_date <= end_date:
                    subject_match = re.search(r'Subject: (.+)', email)
                    if subject_match:
                        sub_match = subject_match.group(1)
    return sub_match

for text in emails :
    nameSenders.append(findSenderName(text))
    mailSenders.append(findSenderMail(text))
    subject_match = findSubject(text, '01 Jun 2002', '13 May 2004')
    if (subject_match.split()):
        subject.append(subject_match)

print("Total Sender Name :  " + str(len(nameSenders)))
print("Total Sender Email : "  + str(len(mailSenders)))
print("Total Subject : "  + str(len(subject)))

# for i in nameSenders:
#     print(i)
    
for i in mailSenders:
    print(i)

# for i in subject:
#     print(i)
    
````
